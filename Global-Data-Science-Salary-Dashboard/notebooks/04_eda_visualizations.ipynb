# 04. Exploratory Data Analysis (EDA) & Visualization

## Goal
Visualize the data to uncover trends and insights. We will focus on answering our key project questions regarding salaries in India vs. the Global market.

## Visualizations
1. **Salary Distribution**: How are salaries distributed globally?
2. **India vs Global Comparison**: The core insight of this project.
3. **Salary by Experience Level**: How does seniority affect pay?
4. **Top Paying Roles**: Which job categories pay the most?
5. **Remote Ratio**: Is remote work common in high-paying roles?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

### 1. Load Processed Data

```python
file_path = "../data/cleaned/processed_jobs_data.csv"
df = pd.read_csv(file_path)
df.head()
```

### 2. Salary Distribution (Global)
Visualizing the overall spread of salaries in USD.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], bins=30, kde=True, color='skyblue')
plt.title('Global Distribution of Data Science Salaries (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

### 3. India vs Global Salary Comparison
This is a critical insight for the project. We expect a gap, but let's visualize the magnitude.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='job_location_type', y='salary_in_usd', palette='Set2')
plt.title('Salary Comparison: India vs Global')
plt.ylabel('Salary (USD)')
plt.xlabel('Location Group')
plt.show()
```

### 4. Salary by Experience Level
Checking salary progression from Entry Level to Executive.

```python
# Order for the plot
order = ['Entry Level', 'Mid Level', 'Senior Level', 'Executive Level']

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='experience_level', y='salary_in_usd', estimator='median', order=order, palette='viridis')
plt.title('Median Salary by Experience Level')
plt.ylabel('Median Salary (USD)')
plt.show()
```

### 5. Top Paying Job Categories
Which roles earn the most on average?

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='salary_in_usd', y='job_category', estimator='mean', palette='magma', ci=None)
plt.title('Average Salary by Job Category')
plt.xlabel('Average Salary (USD)')
plt.show()
```

### 6. Insights Summary
- **Global vs India**: There is likely a significant difference in nominal USD salaries.
- **Experience Premium**: Executive roles command a high premium.
- **Top Roles**: Architect and Lead roles tends to pay higher than standard Analyst roles.
